```python
# 라이브러리 임포트
import tensorflow as tf
import pandas as pd
```

```python
baseball_2015 = pd.read_csv('./baseball_2015.csv', encoding='CP949')
baseball_2015.head()
```

```python
baseball_2016 = pd.read_csv('./baseball_2016.csv', encoding='CP949')
baseball_2016.head()
```

```python
baseball_2017 = pd.read_csv('./baseball_2017.csv', encoding='CP949')
baseball_2017.head()
```

```python
baseball_2018 = pd.read_csv('./baseball_2018.csv', encoding='CP949')
baseball_2018.head()
```

```python
baseball_2019 = pd.read_csv('./baseball_2019.csv', encoding='CP949')
baseball_2019.head()
```

```python
baseball_data = pd.concat([baseball_2015, baseball_2016,baseball_2017, baseball_2018, baseball_2019])
baseball_data
```

```python
# 분석에 필요하다고 생각되는 column으로 df 재구성
baseball_data_re = baseball_data[['T_ID', 'VS_T_ID', 'PA', 'AB', 'RUN', 'OBP', 'OOO', 'win']]
baseball_data_re.head()
```

```python
# 결측치 확인
baseball_data_re.isnull().sum()
```

```python
# baseball_data_re = baseball_data.dropna(axis=0) # 결측치가 없어서 생략
```

```python
baseball_data_re.info() # 데이터 타입과 전체 몇 경기의 데이터인지 확인
```

```python
# 무승부 제거
baseball_data_re = baseball_data_re[baseball_data_re.win != 0.5]
```

```python
# 무승부가 제거된 경기 수 확인: 7288 경기 중 82 경기 (1.125%)
baseball_data_re.info()
```

```python
baseball_data_re['T_ID'].unique()
```

```python
# NE(넥센)은 WO(키움)으로 변경된 동일한 팀
baseball_data_re = baseball_data_re.replace('NE', 'WO')
```

```python
# 팀명 한번 더 확인
baseball_data_re['T_ID'].unique() # NE(넥센)이 없어진 걸 확인할 수 있음
```

```python
# 라벨인코더로 10개 팀 넘버링하기
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(baseball_data_re['T_ID'])
baseball_data_re.loc[:,'T_ID'] = le.transform(baseball_data_re['T_ID'])
baseball_data_re.loc[:,'VS_T_ID'] = le.transform(baseball_data_re['VS_T_ID'])

baseball_data_re.head()
```

```python
baseball_data_re['T_ID'].unique()
```

```python
# # 생략
# # 홈팀(B)은 1로 원정팀(T)는 0으로 변경
# baseball_data = baseball_data.replace('T', 0)
# baseball_data = baseball_data.replace('B', 1)
```

```python
print(baseball_data_re.columns)
```

```python
# 승패에 대한 2진 분류를 위해 원핫인코더로 승패를 2개의 column으로 나눈다.
baseball_data_re = pd.get_dummies(baseball_data_re, columns = ['win'])
baseball_data_re.head()
```

```python
baseball_data_re.dtypes
# "원핫인코더"를 하면서 승패 colunm에 변화가 생김
```

```python
baseball_data_re['win_0.0'] = baseball_data_re['win_0.0'].astype('int')
# #baseball_data['승패_0.5'] = baseball_data['승패_0.5'].astype('int')
baseball_data_re['win_1.0'] = baseball_data_re['win_1.0'].astype('int')
```

```python
# 독립변수, 종속변수
# 위 print(baseball_data.columns)의 index를 이용한다.
data = baseball_data_re[['T_ID', 'VS_T_ID', 'PA', 'AB', 'RUN', 'OBP', 'OOO']]
target = baseball_data_re[['win_0.0', 'win_1.0']] # 무승부는 2진 분류를 위해 위에서 제거하였다.
# target = baseball_data_re[['win']] # 승패 분류가 아닌 회귀로 0 또는 1을 추정해 보려 한다.
print(data.shape, target.shape)
```

```python
# train_x, train_y = train_input, train_target
# test_x, test_y = test_input, test_target

from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(data, target, test_size=0.2, random_state=42)
```

```python
# # train을 이용하여 검증용 sub(진짜 훈련셋)과 val(검증셋)으로 구분하기

# sub_x, val_x, sub_y, val_y = train_test_split(train_x, train_y, test_size=0.2, random_state=42)
```

```python
print(train_x.shape, test_x.shape)
# print(sub_x.shape, val_x.shape)
# # 5764개의 train data가 4611개의 sub data와 1153개의 val data로 나누어졌다.
```

```python
# 모델을 만든다.

X = tf.keras.layers.Input(shape=[7])

H = tf.keras.layers.Dense(10)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(7)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(5)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

Y = tf.keras.layers.Dense(2, activation='softmax')(H)
model = tf.keras.models.Model(X, Y)
model.compile(loss='categorical_crossentropy', metrics='accuracy')
```

```python
# 모델이 잘 만들어졌는지 확인한다.

model.summary()
```

```python
# 훈련이 잘 작동하는지 확인해 본다.
model.fit(train_x, train_y, epochs=10)
```

```python
model.fit(train_x, train_y, epochs=100)
```

```python
model.fit(train_x, train_y, epochs=100, verbose=1)
# "verbose=0"을 이용해서 훈련과정(fit)을 출력하지 않는다.
```

```python
model.fit(train_x, train_y, epochs=100, verbose=1)
```

```python
print(model.predict(test_x[0:10]))
print(test_y[0:10])
```

```python
model.evaluate(test_x, test_y)
```
