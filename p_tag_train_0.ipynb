```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/soyeun-data/p-tag/main/%EC%98%81%ED%99%94%20%EA%B4%80%EA%B0%9D%EC%88%98/movies_train.csv'
df_train = pd.read_csv(url)
```

```python
df_train.head()
```

```python
df_train.info()
```

```python
# dir_prev_bfnum 항목 제거
df = df_train.drop(['dir_prev_bfnum'], axis=1)

# df_train에서 df로 변수명 변경
df
```

```python
df.sort_values('release_time') # 2010년부터 2015년까지 6년간의 영화
```

```python
df_director = df['director'].unique()
len(df_director) # 총 472명의 감독
```

```python
df_distributor = df['distributor'].unique()
len(df_distributor) # 총 169개의 배급사
```

```python
df_genre = df['genre'].unique()
len(df_genre) # 총 12개의 장르
```

```python
df['director'].value_counts() # 472명 감독별 영화수
```

```python
df_distributor_sum = df.groupby('distributor').sum()
df_distributor_sum_sort = df_distributor_sum.sort_values('box_off_num', ascending=False)
```

```python
df_distributor_sum_sort # 169개 배급사별 누적 관객수
```

```python
df['distributor'].value_counts() # 169개 배급사별 영화수
```

```python
df_genre_sum = df.groupby('genre').sum()
df_genre_sum_sort = df_genre_sum.sort_values('box_off_num', ascending=False)
```

```python
df_genre_sum_sort # 12개의 장르별 누적관객수
```

```python
df['genre'].value_counts() # 12개 장르별 영화수
```

```python
df_director_sum = df.groupby('director').sum()
df_director_sum_sort = df_director_sum.sort_values('box_off_num', ascending=False) # 영화감독별 총 관람객수
```
